Classification:
    >target belongs to finite set of classes(categories) 
    exp:gender detection from face,letter prediction,digit prediction,cancer
detection
        >binary classification (2 classes) 
        >multiclass classification(more than 2 classes) 
Regression:
    >target belongs to any real value(int/float) 
    exp:
        predicting fuel price,house price,covid cases till
        december 2020,salary
        >simple regression(one feature in the dataset) 
        >multiple regression(more than one feature in the dataset)
        >polynomial regression(special type of multiple regression)
        
    example of Simply Regression:
        exp     sal 
        1       10k 
        2       12k 
        3       13k 
        5        ? 
    example of Multiple Regression:
        exp      projects     sal 
        1        2            10k
        2        3            12k 
        3        3            13k 
        5        6             ?
        
    example of Polynomial Regression:
    exp      exp*exp     sal 
    1        1            10k
    2        4            12k 
    3        9            13k 
    5        25             ?


```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score,r2_score # here we cannot use accuracy score for regression problem this is for only classification problem
df = pd.read_csv('dataset/regression/salary_1_variable.csv') 
X = df.iloc[:,:-1] 
y = df.iloc[:,-1] 
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
knr = KNeighborsRegressor() # Here it also accecpt all the parameters as like KNeighborsClassifier
knr.fit(X_train, y_train) 
pred_train = knr.predict(X_train)
pred_test = knr.predict(X_test) 
# print(f'accuracy score at : {accuracy_score(y_train,pred_train)}')# we cannot use accuracy score for regression
```

```python
df = pd.read_csv('dataset/regression/salary_1_variable.csv') 
X = df.iloc[:,:-1] 
y = df.iloc[:,-1] 
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
knr = KNeighborsRegressor() # Here it also accecpt all the parameters as like KNeighborsClassifier
knr.fit(X_train, y_train) 
pred_train = knr.predict(X_train)
pred_test = knr.predict(X_test) 
print(f'accuracy score on seen data(training data) : {r2_score(y_train,pred_train)}')# we cannot use accuracy score for regression
print(f'accuracy score on unseen data(testing data) :{r2_score(y_test, pred_test)}')
```

# Internal working of LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression(fit_intercept=True) # <----------- by providing fit_intercept = True may be it will enhance your model
lr.fit(X_train, y_train)
pred_train = lr.predict(X_train)
pred_test = lr.predict(X_test) 

print(f'Training score is : {r2_score(pred_train, y_train)}') 
print(f'Testing score is {r2_score(y_test, pred_test)}')
```

```python
lr.coef_ # co-efficient of x
```

```python
lr.intercept_ # intercept of the equation
```

### If we change the value of fit_intercept=False

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression(fit_intercept=False) # <----------- by providing fit_intercept = True may be it will enhance your model
lr.fit(X_train, y_train)
pred_train = lr.predict(X_train)
pred_test = lr.predict(X_test) 

print(f'Training score is : {r2_score(pred_train, y_train)}') 
print(f'Testing score is {r2_score(y_test, pred_test)}')

import matplotlib.pyplot as plt 
plt.scatter(X_train, y_train,c='black') 
plt.plot(X_train, pred_train,c='red') 
plt.legend(('Out dataset feature matrix and responce vector', 'model prediction at training data'))
plt.show()
```
