```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
```

```python
df = pd.read_csv('Med_Domain.csv') 
df
```

```python
df.cancer.value_counts() # class 1 is very less comparatively to 0 class
```

```python
X = df.drop(['cancer'],axis=1) 
y = df['cancer']
```

```python
# ------------Up-sampling to fix the imbalanced dataset--------------------

ros = RandomOverSampler() 
X_train_res,y_train_res = ros.fit_resample(X,y) 
# ---------------------------------------------------------------------------------------------------------

X_train, X_test, t_train, t_test = train_test_split(X_train_res.values,y_train_res.values, test_size=0.3, shuffle=True, random_state=1)
```
