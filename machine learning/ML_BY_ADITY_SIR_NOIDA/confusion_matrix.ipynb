```python
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import confusion_matrix 
```

```python
df = pd.read_csv('/home/echelon/machine_learning_personal/machine learning/dataset/classification/fruits.csv') 
df
```

```python
df.FruitName = df.FruitName.map({'Apple':0,'Banana':1})
X = df.iloc[:,:-1]
y = df.iloc[:,-1] 
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=10) 

knc = KNeighborsClassifier()
knc.fit(X_train,y_train)
pred_train = knc.predict(X_train) 
pred_test = knc.predict(X_test)
```

```python
cm_train = confusion_matrix(y_train,pred_train)
cm_test = confusion_matrix(y_test,pred_test)
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_train,pred_train) # You can see the same accuracy value
```

```python
from sklearn.metrics import recall_score,precision_score
recall_score(y_train,pred_train) # bydefault it gives the 1-class  score 
```

```python
recall_score(y_train,pred_train, average=None) # to see all the class recall
```

```python
precision_score(y_train,pred_train) # by default it gives the 1-class precision_score
```

```python
precision_score(y_train,pred_train,average=None) # to see both the class precision_score
```
