# Practice on diabetes dataset

```python
from sklearn.datasets import load_diabetes 
import pandas as pd 
```

```python
bunch_obj = load_diabetes() 
X = bunch_obj['data']  
X
```

```python
y = bunch_obj['target'] 
y
```

```python
df = pd.DataFrame(X, columns=bunch_obj['feature_names']) 
df
```

```python
df['y'] = y # Here y column is the target/label column
df
```

# correlation

```python
df.corr()   # It returna pivot table
```

```python
import seaborn as sns
```

```python
pivot_table = df.corr() 
sns.heatmap(pivot_table) # this is the effctive way to visualise the correlation b/w the two things/features using the heatmap
```

```python
df.drop('sex', axis=1,inplace=True) # axis : {0 or 'index', 1 or 'columns'}, default 0
df
```

```python
X = df.iloc[:,:-1].values 
y = df.iloc[:,-1]  
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10) 
```

```python
from sklearn.linear_model import LinearRegression 
lr = LinearRegression() 
lr.fit(X, y) 
print(f'Training Score : {lr.score(X_train, y_train)}')
print(f'Testing Score : {lr.score(X_test, y_test)}')
```
