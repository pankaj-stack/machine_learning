```python
X=[]
y=[]
```

```python
import os
import numpy as np

for folder_name in ['pankaj','aman']:
    img_names=os.listdir(f"{video_path}/train/{folder_name}")
    for name in img_names:
        img=cv2.imread(f"{video_path}/train/{folder_name}/{name}")
        img=cv2.resize(img,(100,100))
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        gray=gray.flatten()
        gray=gray/255 # Scaling of the dataset
        X.append(gray)
        y.append(folder_name.upper())

np.shape(X)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV 
from sklearn.metrics import classification_report
import warnings
```

```python
%%time 
warnings.filterwarnings('ignore')

grid_parameters = [{ "penalty": ['l2'],
                     "tol":[0.001,0.0001],
                     "C":[1,2,3],
                     'max_iter':[100,150,200]}]

log=LogisticRegression()
grid = GridSearchCV(log,grid_parameters)
grid.fit(X,y) 

y_pred_train = grid.predict(X)
display(classification_report(y,y_pred_train))
display(grid.best_params_)
display(grid.best_score_)
```
