# --------------------Outlier removal from the dataset--------------------

```python
def drop_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   not_outliers = df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   outliers_dropped = not_outliers.dropna().reset_index()

   return outliers_dropped

outlier_removed_df = drop_outliers_IQR(df) 
outlier_removed_df
```

# --------------without scaling----------------

```python
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
df = outlier_removed_df
X = df.iloc[:,:-1] 
y = df.iloc[:,-1] 

X_train,X_test,y_train,y_test = train_test_split(X,y)  

knc = KNeighborsClassifier() 
knc.fit(X_train,y_train) 

print(f'Score at training data : {knc.score(X_train,y_train)}')
print(f'Score at testing data : {knc.score(X_test,y_test)}')

print(f'{classification_report(knc.predict(X_test),y_test)}')

```

# ----------------Scaling the dataset by StandardSacaler----------------

```python
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
df = outlier_removed_df
X = df.iloc[:,:-1] 
y = df.iloc[:,-1] 

X_train,X_test,y_train,y_test = train_test_split(X,y)  

scaler = StandardScaler() # yhan scaling se accuracy score badcha hai 
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

knc = KNeighborsClassifier() 
knc.fit(X_train,y_train) 

print(f'Score at training data : {knc.score(X_train,y_train)}')
print(f'Score at testing data : {knc.score(X_test,y_test)}')

print(f'{classification_report(knc.predict(X_test),y_test)}')

```
