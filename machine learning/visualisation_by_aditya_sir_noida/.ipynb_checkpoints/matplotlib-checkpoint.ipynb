### marker 

```python
overs = [1,2,3,4,5] 
runs = [10,8,15,20,10] 
markers_list = ['.',',','o','v','^','<','>','1','2','3','4','s','p','*','h','H','+','x','D','d','|','_']
for markers in markers_list:
    print(markers)
    plt.plot(overs,runs,marker=markers)
    plt.xlabel('Overs') 
    plt.ylabel('Runs') 
    plt.xticks([1,2,3,4,5])
    plt.show()
```

```python
overs = [1,2,3,4,5] 
runs = [10,8,15,20,10] 
plt.plot(overs,runs,marker='s',c='yellow')
plt.xlabel('Overs') 
plt.ylabel('Runs') 
for x,y in zip(overs,runs):
    plt.text(x,y,f'{x} over {y} runs',size=15,c='blue') # <----------
plt.title('Overs/Runs')
plt.show()         
```

```python
overs = [1,2,3,4,5] 
runs = [10,8,15,20,10] 
runs2 = [5,10,15,20,25]
plt.plot(overs,runs,marker='s',c='yellow')
plt.plot(overs,runs2,marker='o',c='g')
plt.xlabel('Overs') 
plt.ylabel('Runs') 
for x,y in zip(overs,runs):
    plt.text(x,y,f'{x} over {y} runs',size=15,c='blue') # <----------
plt.title('Team1 vs Team2')
plt.legend(['Team1','Team2'])
plt.show()         
```

```python
overs = [1,2,3,4,5] 
runs = [10,18,33,53,63] 
runs2 = [5,15,16,36,61]
plt.figure(figsize = (7,10)) # <---------
plt.plot(overs,runs,marker='s',c='yellow')
plt.plot(overs,runs2,marker='o',c='g')
plt.xlabel('Overs') 
plt.ylabel('Runs') 
for x,y in zip(overs,runs):
    plt.text(x,y,y,size=15,c='blue') 

for x,y in zip(overs,runs2):
    plt.text(x,y,y,size=15,c='blue') 

plt.title('Team1 vs Team2')
plt.legend(['Team1','Team2'])
plt.grid() # <----------
plt.show()         
```

```python
overs = [1,2,3,4,5] 
runs = [10,18,33,53,63] 
runs2 = [5,15,16,36,61]
plt.plot(overs,runs,marker='s',c='yellow')
plt.plot(overs,runs2,marker='o',c='g')
plt.xlabel('Overs') 
plt.ylabel('Runs') 
for x,y in zip(overs,runs):
    plt.text(x,y,y,size=15,c='blue') 

for x,y in zip(overs,runs2):
    plt.text(x,y,y,size=15,c='blue') 

plt.title('Team1 vs Team2')
plt.legend(['Team1','Team2'])
plt.savefig('Team1 vs Team2') # <-------on the current working directory if did not provide the path
plt.show()         
```
