```python
import statistics as stats 
x = [4,6] 
print(stats.mean(x)) # ye whan use hota hai jhan range same hoti hai values ki
print(stats.harmonic_mean(x))# ye whan use hota hai jhan rate diya hota ,jhan 2 unit ki baat ho rhi hogi jaise km/hr
print(stats.geometric_mean(x))# ye whan use hota hai jhan values are in different range me hoti jaise insaan ki height and tree ki hight

```

```python
x = [1,2,1,3,1,4,2,1,3,4] 
print(stats.mode(x))# mode is the most frequent value i.e 1
```
