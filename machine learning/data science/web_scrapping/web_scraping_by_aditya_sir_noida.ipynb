```python
#practice
import requests
resp=requests.get('https://www.mohfw.gov.in/')
resp.content#It returns the data into bytes whichi is used to read image
resp.text#It returns data as a string
import BeautifulSoup
data=resp.text
soup=BeautifulSoup(data,'parser')
soup.prettify()#Formatting of html
```

```python
page_text=soup.getText()#How to extract text of complete web page
print(page_text)
```

```python
page_text=soup.getText(strip=True)
print(page_text)
```

```python
page_text=soup.getText(strip=True,separator='\n')#Here separator='\n' means to print all the text in the new  line
print(page_text)
```

```python
#How To sava the data into the storage
with open('mohfw.txt','w',encoding='utf8') as f:#Here you can also use 'utf' encoding
    f.write(page_text)
    f.close()
print('file is copied into the memory')
```

```python
#How to get title tag
title=soup.find('title')
print(type(title))
print(title)#MoHFW | Home <---------text of the tag
```

```python
#How to get link(anchor tag)
anchor=soup.find('a')
print(type(anchor),anchor,sep='\n')
print(anchor.getText())#To get the text of the anchor tag
```

```python
#How to get the attribute of the anchor
print(anchor.get('href'))#It works like dictionary
```

```python
#How to get all links(anchor tags)
anchors=soup.find_all('a')#Returns list of all tags
print('No. of links',len(anchors))
for anchor in anchors:
    print(anchor)
```

```python
import os
anchors=soup.find_all('a')#Returns list of all tags
file=open('all_links.txt','w',encoding='utf8')
for anchor in anchors:
    file.write(str(anchor))
file.close()
print(f'File is saved at {os.getcwd()} position')
```

```python
#How to get all links (anchor tags) whose href value starts with https
anchor=soup.find_all('a')
https_anchors=[]
for anchor in anchors:
    href_value=anchor.get('href')
    if href_value.startswith('https'):
        https_anchors.append(anchor)
        print(anchor)
print('Total links whose href value starts with https : ',len(https_anchors))
```
