```python
import matplotlib.pyplot as plt
import numpy as np
import random
ml_students_age=np.random.randint(18,45,(100))
py_students_age=np.random.randint(15,40,(100))
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=.8,histtype='bar',color='magenta',
         orientation='horizontal',label='ML Students')
plt.title("ML Student's age graph",fontsize=20)
plt.ylabel("Student's age category")
plt.xlabel('No. of students')
plt.legend(loc='best')
plt.show()
```

```python
#Plotting both the graphs simultaneously
import matplotlib.pyplot as plt
import numpy as np
import random
ml_students_age=np.random.randint(18,45,(100))
py_students_age=np.random.randint(15,40,(100))
bins=[15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age],bins,rwidth=.8,histtype='bar'
         ,color=['m','g'],orientation='vertical',
         label=['ML Students','Python Students'])
plt.title("ML Student's age graph",fontsize=20)
plt.xlabel("Student's age category")
plt.ylabel('No. of students')
plt.legend(loc='best')
plt.show()
```

```python
#Plotting both the graphs simultaneously
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
style.use('ggplot')
ml_students_age=np.random.randint(18,45,(100))
py_students_age=np.random.randint(15,40,(100))
bins=[15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age],bins,rwidth=.8,histtype='bar',color=['m','y'],
         orientation='vertical',label=['ML Students','Python Students'])
plt.title("ML Student's age graph",fontsize=20)
plt.ylabel("Student's age category")
plt.xlabel('No. of students')
plt.legend(loc='best')
plt.show()
```
