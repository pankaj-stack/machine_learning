```python
# ## Matplotlib Visualisation Methods or Fuctions
# 1.np.arange(starting_point,end_point,step)#It creates the n-di-arry
# 2.DataFrame()<----------pandas
# 2.plot(x,y)#to plot the given points on the graph
# 3.show()# to display the  plotted graph
# 4.marker#to show  points on the curve
# 5.markersize#size of the marker
# 6.linewidth#width of the curve/line
# 7.linestyle# style of line/curve
# 8.color#to give the color to your curve
# 9.alpha#if set the contrast of the line/curve
# 10.xlabel()#to write the name of x-axis
# 11.ylabel()#to write the name of y-axis
# 12.title()# to write the name of the graph
# 13.xticks(it takes the sequence)#to plot the points on the x-axis manually
# 14.yticks(it takes the sequence)#to plot the points on the y-axis manually
# 15.xlim()# it cut the required x-axis portion
# 16.ylim()# it cut the required y-axis portion
# 17.legend()
```

```python
# Matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#%matplotlib inline#optional 
```

```python
#y=x^2
x=np.arange(10)
y=x**2
pd.DataFrame([x,y])
```

```python
plt.plot(x,y)
plt.show()#It works as print function to display the graph
```

```python
plt.plot(x,y,marker='o',markersize=10)
plt.show()
```

```python
plt.plot(x,y,linewidth=3,linestyle='--',color='yellow',alpha=0.8,marker='o',markersize=10)
plt.show()
```

```python
plt.plot(x,y,label='x^2')
plt.xlabel('------x axis-------->')
plt.ylabel('------y axis--------->')
plt.title('y=x^2 Graph')
plt.xticks(range(10))
plt.yticks([0,50,80])
plt.xlim([3,6])
plt.ylim([8,40])
plt.legend()#special notes on the axises
plt.show()
```

```python
# Sin curve,y=sin(x)
series=np.linspace(-np.pi,(np.pi)+1)# it gives us 50 equal division by default in the given interval
sine=np.sin(series)
plt.plot(sine)
plt.show()
```

# Bar Plotting

```python
weekly_data=weather_df.groupby(weather_df.index.dayofweek).mean()
weekly_data
```

```python
plt.bar(weekly_data.index,weekly_data['Visibility (km)'])
plt.xlabel('Weeks')
plt.ylabel('Mean visibility')
plt.xticks(weekly_data.index,calendar.day_abbr,rotation=45)
plt.show()
```

```python
plt.barh(weekly_data.index,weekly_data['Visibility (km)'])
plt.xlabel('Mean visibility')
plt.ylabel('Weeks')
plt.title('weekly visibility trend')
plt.yticks(weekly_data.index,calendar.day_abbr,rotation=45)
plt.show()
```

```python
plt.bar(weekly_data.index-0.2,weekly_data['Visibility (km)'],width=0.4,
       label='visibility',color='g')
plt.bar(weekly_data.index+0.2,weekly_data['Wind Spd (km/h)'],width=0.4,
       label='wind_speed')
plt.xlabel('Weeks')
plt.ylabel('Mean Values')
plt.title('weekly trend')
plt.xticks(weekly_data.index,calendar.day_abbr,rotation=45)
plt.legend(loc='best')
plt.figure(figsize=(8,7))
plt.show()
plt.
```

```python
#From the internet
N = 5
men_means = (20, 35, 30, 35, 27)
men_std = (2, 3, 4, 1, 2)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, men_means, width, color='r', yerr=men_std)

women_means = (25, 32, 34, 20, 25)
women_std = (3, 5, 2, 3, 3)
rects2 = ax.bar(ind + width, women_means, width, color='y', yerr=women_std)

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))

ax.legend((rects1[0], rects2[0]), ('Men', 'Women'))


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```
