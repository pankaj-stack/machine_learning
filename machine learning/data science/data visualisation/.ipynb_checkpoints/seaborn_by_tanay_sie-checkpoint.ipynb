```python
data=sns.load_dataset('tips')
```

```python
abc=sns.load_dataset('tips')
sns.swarmplot(x='tip',y='smoker',data=abc)
```

```python
titanic=sns.load_dataset('titanic')
g=sns.factorplot('class','survived','sex',data=titanic,kind='bar',legend=True)
plt.show()
```

```python
import numpy as np
data=np.arange(1,49,1).reshape(12,4)
print(data)
sns.boxplot(data=data)
plt.show()
```

```python
df=sns.load_dataset('tips')
sns.violinplot(x='day',y='tip',hue='smoker',data=df,palette='Pastel1')
plt.show()
```

```python
df=sns.load_dataset('tips')
sns.set
sns.set_style('whitegrid')
sns.swarmplot(x='tip',y='smoker',data=df)
plt.plot()
```
