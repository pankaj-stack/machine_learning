```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
classes=['Python','R','AI','ML','DS']
class1_students=[30,10,20,25,10]
class2_students=[40,5,20,20,10]
class3_students=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
plt.show()
```

```python
plt.bar(classes,class1_students,color='y',edgecolor='m',linewidth=5,alpha=.9,
       linestyle='--',align='edge',width=.3)
plt.show()
```

```python
plt.bar(classes,class1_students,color='y',edgecolor='m',linewidth=5,alpha=.9,
       linestyle='--',align='edge',width=.3,visible=False)#visible == False means graph is not going show
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(10,6))
plt.bar(classes,class1_students,color='k',edgecolor='m',linewidth=5,alpha=.9,
       linestyle='--',align='edge',width=.3)
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(10,6))

plt.bar(classes,class1_students,color='k',width=.3,label='Class1 students')
plt.bar(classes,class2_students,color='g',width=0.2,label='Class2 Students')

plt.title('Bar Chart of IAIP Classes',fontsize=18)
plt.xlabel('Classes',fontsize=13)
plt.ylabel('No. of Students',fontsize=13)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(10,6))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,color='k',width=width,label='Class1 students')
plt.bar(classes_index+width,class2_students,color='g',width=width,label='Class2 Students')

plt.title('Bar Chart of IAIP Classes',fontsize=18)
plt.xlabel('Classes',fontsize=13)
plt.ylabel('No. of Students',fontsize=13)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
style.use('ggplot')
plt.figure(figsize=(10,6))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,color='k',width=width,label='Class1 students')
plt.bar(classes_index+width,class2_students,color='g',width=width,label='Class2 Students')

plt.title('Bar Chart of IAIP Classes',fontsize=18)
plt.xlabel('Classes',fontsize=13)
plt.ylabel('No. of Students',fontsize=13)
plt.xticks(classes_index+width,classes,rotation=45)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(10,6))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,color='k',width=width,label='Class1 students')
plt.bar(classes_index+width,class2_students,color='g',width=width,label='Class2 Students')
plt.bar(classes_index+width+width,class3_students,color='m',width=width,label='Class3 Students')

plt.title('Bar Chart of IAIP Classes',fontsize=18)
plt.xlabel('Classes',fontsize=13)
plt.ylabel('No. of Students',fontsize=13)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
style.use('ggplot')
classes_index=np.arange(len(classes))
width=0.2
plt.barh(classes_index,class1_students,color='k',label='Class1 students')
plt.barh(classes_index+width,class2_students,color='g',label='Class2 Students')
plt.barh(classes_index+width+width,class3_students,color='m',label='Class3 Students')

plt.title('Bar Chart of IAIP Classes',fontsize=18)
plt.xlabel('Classes',fontsize=13)
plt.ylabel('No. of Students',fontsize=13)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
