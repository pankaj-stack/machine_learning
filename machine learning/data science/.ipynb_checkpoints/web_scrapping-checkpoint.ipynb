```python
with open('/home/pankaj/Downloads/Sample.html','at') as file:
    for tag in soup.find_all('div',{'class':'article'}):
        heading=tag.h2.a.text
        summary=tag.p.text
        file.write(f'{heading}:{summary}')#it always write string type
        file.write('\n')
    print('Recorded Successfully')
```

```python
with open('/home/pankaj/Downloads/Sample.csv','at') as file:
    for tag in soup.find_all('div',{'class':'article'}):
        heading=tag.h2.a.text
        summary=tag.p.text
        file.write(f'{heading}:{summary}')#it asways write string type
        file.write('\n')
    print('Recorded Successfully')
```

```python
with open('/home/pankaj/Downloads/Sample_crap_head.csv','at') as file:
    file.write(f'heading,summary')#it asways write string type
    file.write('\n')
    for tag in soup.find_all('div',{'class':'article'}):
        heading=tag.h2.a.text
        summary=tag.p.text
        file.write(f'{heading},{summary}')#it asways write string type
        file.write('\n')
print('Recorded Successfully')
```

```python
import csv
with open('/home/pankaj/Downloads/Sample_crap_head.csv','at') as file:
    obj=csv.writer(file)
    obj.writerow(['headline','summary'])
    for tag in soup.find_all('div',{'class':'article'}):
        heading=tag.h2.a.text
        summary=tag.p.text
        obj.writerow([heading,summary])
    print('Recorded Successfully')
```

## heading

```python
for head in soup.find_all('h2'):
    print(head.name+' '+head.text)
```

```python
for head in soup.find_all('h2'):
    if ':' in head.text:
        print(head.name+' '+head.text[:((head.text).index(':'))])
    elif '' in head.text:
        print(head.name+' '+head.text[:((head.text).index(''))])
    else:
        print(head.name+' '+head.text)
```

## Summary

```python
for summary in soup.find_all('div'):
    if summary.summary:
        print(summary)
    else:
        print('No summary is there')
        break
```
